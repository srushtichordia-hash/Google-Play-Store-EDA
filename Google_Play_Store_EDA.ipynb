```python
df_copy['Size'].unique()
```

```python
def size_process(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace('M',""))
        res = res*1024
        return res
    elif str(item)[-1] == 'K':
        res = float(str(item).replace('k',""))
        return res
    else:
        return str(np.nan)
        
```

```python
size_process('19M')
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```

```python
df_copy['Size'] = df_copy['Size'].astype(float)
```

# Machine Learning Model

### Target Variable is Rating
### Predicts app rating from features like Reviews, Installs, Size, Category, Type, Price, etc.

Drop high-cardinality columns:

```python
df_copy = df_copy.drop(['App', 'Current Ver', 'Android Ver'],axis=1)
```

```python
df_copy = pd.get_dummies(df_copy,columns=['Category', 'Type', 'Content Rating', 'Genres'],drop_first=True)
```

```python

x = df_copy.drop('Rating',axis = 1)
y = df_copy['Rating']
```

```python
y.isna().sum()
y.fillna(y.mean(),inplace = True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 100,random_state = 42)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
y_pred = model.predict(x_test)
print("R2 Score:", r2_score(y_test,y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test,y_pred))
```

# Feature Importance Plot

```python
importance = pd.Series(model.feature_importances_,index=x.columns).sort_values(ascending=False)

plt.figure(figsize=(10,5))
importance.head(10).plot(kind='bar',color = 'Purple')
plt.title("Top Features Affecting Rating")
plt.show()
```
